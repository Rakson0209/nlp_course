````python
from bs4 import BeautifulSoup as bs
import json
import jsonlines
from tqdm import tqdm
````

````python
file = open('./data/med.cd.az.xml', 'r', encoding = 'utf-8')
content = file.readlines()
bs_contents = []
for line in tqdm(content):
    bs_contents.append((bs(line, features="xml")))
print(len(bs_contents))
````

````python
word_dict = {}
for bs_content in tqdm(bs_contents):
    for entry in bs_content.find_all('HOMOGRAPH'):
        # tip1: view what tags each entry may contain before proceeding.
        # tip2: not all entries have all the attributes. BeautifulSoup returns `None` when it finds nothing.
        # Note that some tags might not look exactly like those in the output example 
            # (e.g. SENSE45, TRANSLATION). 
            # So please remember to change the tag names as required when you save them to your own dict.
        key = ''
        if entry.find('ENTRY0') != None:
            key = entry.find('ENTRY0').get_text()
            if key in word_dict:
                key = entry.find('ENTRY0').get_text() + '_1'
        elif entry.find('ENTRY13') != None:
            key = entry.find('ENTRY13').get_text()
            if key in word_dict:
                key = entry.find('ENTRY13').get_text() + '_1'
        
        content = {}
        #ID
        content['ID'] =  entry['ID']
        #PART-OF-SPEECH
        if entry.find('PART-OF-SPEECH') != None:
            content['PART-OF-SPEECH'] = entry.find('PART-OF-SPEECH').get_text()
        if entry.find('SENSE45') != None:
            sense_list = []
            for sense in entry.find_all('SENSE45'):
                sense_dict = {}
                # DEFINITION
                if sense.find('DEFINITION') != None:
                    sense_dict['DEFINITION'] = sense.find('DEFINITION').get_text()
                if sense.find('trans') != None:
                    sense_dict['TRANSLATION'] = sense.find('trans').get_text()
                if sense.find('EXAMPLES') != None:
                    example_list = []
                    for example in sense.find_all('EXAMPLES'):
                        example_dict = {}
                        if example.find('EXAMPLE') != None:
                            example_dict['EXAMPLE'] = example.find('EXAMPLE').get_text()
                        if example.find('trans') != None:
                            example_dict['TRANSLATION'] = example.find('trans').get_text()
                        example_list.append(example_dict)
                    sense_dict['EXAMPLES'] = example_list
                if sense.find('SUB-SENSE') != None:
                    sub_sense_list = []
                    for sub_sense in sense.find_all('SUB-SENSE'):
                        sub_sense_dict = {}
                        if sub_sense.find('DEFINITION') != None:
                            sub_sense_dict['DEFINITION'] = sub_sense.find('DEFINITION').get_text()
                        if sub_sense.find('trans') != None:
                            sub_sense_dict['TRANSLATION'] = sub_sense.find('trans').get_text()
                        if sub_sense.find('EXAMPLES') != None:
                            sub_sense_example_list = []
                            for sub_sense_example in sub_sense.find_all('EXAMPLES'):
                                sub_sense_example_dict = {}
                                if sub_sense_example.find('EXAMPLE') != None:
                                    sub_sense_example_dict['EXAMPLE'] = sub_sense_example.find('EXAMPLE').get_text()
                                if sub_sense_example.find('trans') != None:
                                    sub_sense_example_dict['TRANSLATION'] = sub_sense_example.find('trans').get_text()
                                sub_sense_example_list.append(sub_sense_example_dict)
                            sub_sense_dict['EXAMPLES'] = sub_sense_example_list
                        sub_sense_list.append(sub_sense_dict)
                    sense_dict['SUB-SENSE'] = sub_sense_list
                sense_list.append(sense_dict)
            content['SENSE45'] =sense_list
                
        if entry.find('SENSE') != None:
            sense_list = []
            for sense in entry.find_all('SENSE'):
                sense_dict = {}
                # DEFINITION
                if sense.find('DEFINITION') != None:
                    sense_dict['DEFINITION'] = sense.find('DEFINITION').get_text()
                if sense.find('trans') != None:
                    sense_dict['TRANSLATION'] = sense.find('trans').get_text()
                if sense.find('EXAMPLES') != None:
                    example_list = []
                    for example in sense.find_all('EXAMPLES'):
                        example_dict = {}
                        if example.find('EXAMPLE') != None:
                            example_dict['EXAMPLE'] = example.find('EXAMPLE').get_text()
                        if example.find('trans') != None:
                            example_dict['TRANSLATION'] = example.find('trans').get_text()
                        example_list.append(example_dict)
                    sense_dict['EXAMPLES'] = example_list
                if sense.find('SUB-SENSE') != None:
                    sub_sense_list = []
                    for sub_sense in sense.find_all('SUB-SENSE'):
                        sub_sense_dict = {}
                        if sub_sense.find('DEFINITION') != None:
                            sub_sense_dict['DEFINITION'] = sub_sense.find('DEFINITION').get_text()
                        if sub_sense.find('trans') != None:
                            sub_sense_dict['TRANSLATION'] = sub_sense.find('trans').get_text()
                        if sub_sense.find('EXAMPLES') != None:
                            sub_sense_example_list = []
                            for sub_sense_example in sub_sense.find_all('EXAMPLES'):
                                sub_sense_example_dict = {}
                                if sub_sense_example.find('EXAMPLE') != None:
                                    sub_sense_example_dict['EXAMPLE'] = sub_sense_example.find('EXAMPLE').get_text()
                                if sub_sense_example.find('trans') != None:
                                    sub_sense_example_dict['TRANSLATION'] = sub_sense_example.find('trans').get_text()
                                sub_sense_example_list.append(sub_sense_example_dict)
                            sub_sense_dict['EXAMPLES'] = sub_sense_example_list
                        sub_sense_list.append(sub_sense_dict)
                    sense_dict['SUB-SENSE'] = sub_sense_list
                sense_list.append(sense_dict)
            content['SENSE'] =  sense_list
            
        if entry.find('PHRASE') != None:
            phrase_list = []
            for phrase in entry.find_all('PHRASE'):
                phrase_dict = {}
                if phrase.find('DEFINITION') != None:
                    phrase_dict['DEFINITION'] = phrase.find('DEFINITION').get_text()
                if phrase.find('MULTIWORD') != None:
                    phrase_dict['MULTIWORD'] = phrase.find('MULTIWORD').get_text()
                if phrase.find('trans') != None:
                    phrase_dict['TRANSLATION'] = phrase.find('trans').get_text()
                phrase_list.append(phrase_dict)
                if phrase.find('EXAMPLES') != None:
                    phrase_example_list = []
                    for phrase_example in phrase.find_all('EXAMPLES'):
                        phrase_example_dict = {}
                        if phrase_example.find('EXAMPLE') != None:
                            phrase_example_dict['EXAMPLE'] = phrase_example.find('EXAMPLE').get_text()
                        if phrase_example.find('trans') != None:
                            phrase_example_dict['TRANSLATION'] = phrase_example.find('trans').get_text()
                        phrase_example_list.append(phrase_example_dict)
                    phrase_dict['EXAMPLES'] = phrase_example_list
            content['PHRASE'] =  phrase_list
        
        word_dict[key]= content
````
